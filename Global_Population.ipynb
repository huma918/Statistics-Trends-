## Correlation Matrix : HeatMap

```python
# Correlation Matrix
def plot_correlation_matrix(df):
    """
    Plot a correlation matrix heatmap with a larger figure size and clearer annotations.
    """
    numeric_df = df.select_dtypes(include=['float64', 'int64'])
    plt.figure(figsize=(12, 10))
    correlation = numeric_df.corr()
    sns.heatmap(correlation, annot=True, cmap="coolwarm", square=True, fmt=".2f", annot_kws={"size": 8})
    plt.title("Correlation Matrix Heatmap")
    plt.show()
plot_correlation_matrix(data)
```

## Bar Chart

```python
def plot_bar_chart(df, category_column, value_column, interval=10):
    """
    Plot a bar chart with larger intervals for better readability.
    """
    plt.figure(figsize=(14, 8))
    subset_data = df.iloc[::interval]  # Select every nth row based on interval
    sns.barplot(x=category_column, y=value_column, data=subset_data)
    plt.title(f"Bar Chart of {value_column} by {category_column} (interval={interval})")
    plt.xlabel(category_column)
    plt.ylabel(value_column)
    plt.xticks(rotation=45, ha="right")
    plt.show()
plot_bar_chart(data, 'place', 'pop2023', interval=10)
```

## Scatter Plot:

```python
# Scatter Plot
def plot_scatter(df, x_column, y_column):
    """
    Plot a scatter plot with larger intervals on axes for better readability.
    """
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=x_column, y=y_column, data=df)
    plt.title(f"Scatter Plot of {y_column} vs {x_column}")
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.xticks(ticks=range(0, int(df[x_column].max()), int(df[x_column].max() / 5)))
    plt.yticks(ticks=range(0, int(df[y_column].max()), int(df[y_column].max() / 5)))
    plt.show()
plot_scatter(data, 'pop1980', 'pop2023')
```
